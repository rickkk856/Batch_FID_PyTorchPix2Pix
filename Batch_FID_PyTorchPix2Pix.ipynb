```python
#Define Variables

EXPERIMENT_NAME = str("08K_Rplan_Black_256") #Change this according to your model name used on training
Epoch_range = int("200") #Change this according to the number of epochs used on training
Epoch_range = int(Epoch_range // 5)
```

```python
#Code bellow to create folders and generate codes to copy and run on the next cell 
#The codes to run on CMD move files with patterns "_fake_B" and "_real_B" to generated folders
#The last code calculate FID for each EPOCH

import os
from os import listdir
from os.path import isfile, join
import warnings

warnings.filterwarnings("ignore", category=FutureWarning) #Ignore few warnings

EPOCH_COUNTER = 0
EPOCH_COUNTER2 = f'_{EPOCH_COUNTER}'

for i in range(Epoch_range):
    EPOCH_COUNTER += 5
    EPOCH_COUNTER2 = f'_{EPOCH_COUNTER}'
    
    mypath = str("./results/" + EXPERIMENT_NAME + "/test" + EPOCH_COUNTER2 + "/images")
    patterns = ['_fake_B', '_real_B']
    folders = [os.path.join(mypath, pattern) for pattern in patterns]

    # build folders
    for folder in folders:
        if not os.path.exists(folder):
            os.makedirs(folder)

    print("#----------------EPOCH------------------------>",EPOCH_COUNTER)
    print("!mmf move --source ./results/" + EXPERIMENT_NAME + "/test" + EPOCH_COUNTER2 + "/images --dest-dir ./results/" + EXPERIMENT_NAME + "/test" + EPOCH_COUNTER2 + "/images/_real_B --pattern _real_B")
    print("!mmf move --source ./results/" + EXPERIMENT_NAME + "/test" + EPOCH_COUNTER2 + "/images --dest-dir ./results/" + EXPERIMENT_NAME + "/test" + EPOCH_COUNTER2 + "/images/_fake_B --pattern _fake_B")
    print("#----------------FID FOR------------------------>", EPOCH_COUNTER)
    print("print('FID for epoch:", EPOCH_COUNTER , "')")
    print("!python -m pytorch_fid " + "./results/" + EXPERIMENT_NAME + "/test" + EPOCH_COUNTER2 + "/images/_real_B ./results/" + EXPERIMENT_NAME + "/test" + EPOCH_COUNTER2 + "/images/_fake_B")
    print("")
    print("")

 
```
